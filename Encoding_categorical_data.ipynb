## **Encoding categorical data**


convert categorical variables into numbers

```python
import numpy as np
import pandas as pd
```

```python
#country , age and salary
data = [['India',np.nan,68000.0],
    ['France',43.0,45000.0],
 [np.nan,30.0, 54000.0],
 ['France' ,48.0, 65000.0],
 ['Germany' ,40.0, np. nan],
 ['India' ,35.0, 58000.0],
 ['Germany', np.nan ,53000.0],
 ['France' ,49.0, 79000.0],
 ['India', 50.0 ,88000.0],
 [np.nan ,37.0, np.nan]]
```

```python
df = pd.DataFrame(data)
df
```

```python
df[0]
```

```python
df[0].unique()
```

```python
#Replace method
df[0].replace({'India':0, 'France':1, 'NaN':2, 'Germany':3})
```

```python
#Map method
df[0].map({'India':0, 'France':1, 'NaN':2, 'Germany':3})
```

Sklearn preprocessing
# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
lb.fit_transform(df[0])
```

```python
df[0].values
```

```python
lb.classes_
```
